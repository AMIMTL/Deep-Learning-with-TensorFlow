```python
import numpy as np
import pandas as pd

%matplotlib widget
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")

import seaborn as sns
sns.set_style("darkgrid")

%config InlineBackened.format_figure = "svg"
```

```python
# make_blobs is a very useful function to create a dataset
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=300,
                  n_features=2,
                  centers=2,
                  cluster_std=1.0,
                  random_state=101)
type(data)
```

```python
len(data)
```

```python
X , y = data
```

```python
y
```

```python
rng = np.random.default_rng(seed = 101)
z_noise = rng.normal(loc = 0 , scale=1, size=len(X))
z_sereis = pd.Series(z_noise)
z_sereis
```

```python
features = pd.DataFrame(X)
features
```

```python
new_features = pd.concat([features , z_sereis] , axis=1)
new_features.columns = ['X1', 'X2', 'X3']
new_features
```

```python
np.unique(y)
```

```python
plt.scatter(new_features['X1'], new_features['X2'], c = y, cmap = 'PiYG')
plt.title('2D-scatterplot of X1 & X2')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d  import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(new_features['X1'], new_features['X2'], new_features['X3'], s=20, c=y , cmap ='PiYG')
plt.show()
```

```python
import tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
# 3 ---> 2 ---> 3
```

```python
encoder = Sequential()
encoder.add(Dense(units=2 , activation='relu', input_shape = [3]))
```

```python
decoder = Sequential()
decoder.add(Dense(units=3 , activation='relu', input_shape = [2] ))
```

```python
autoencoder = Sequential([encoder , decoder])
autoencoder.compile(loss ="mse", optimizer=SGD(learning_rate=1.5))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(new_features)
```

```python
autoencoder.fit(scaled_data , scaled_data , epochs=100)
```

```python
encoded_2dim = encoder.predict(scaled_data)
```
